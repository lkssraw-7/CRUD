```python
database = []

def create():
  cad = input("Você já possui cadastro? S/N - ")
  if cad.upper() == 'N':
     cad2 =input("Deseja realizar seu cadastro? 'N' para entrar como visitante. S/N")
     if cad2.upper() == 'S':
        nome = input('Digite seu nome:')
        while True:
            email = input('Digite o seu email').strip().lower()
            if '@' in email and '.' in email:
              break
            else:
              print('Email inválido. Tente novamente.')
        senha = input('Digite sua senha: ')
        id = len(database) + 1

        usuario = {
            'email':email,
            'senha':senha,
            'id':id,
            'nome':nome
        }

        database.append(usuario)
        print('Cadastro realizado com sucesso!')
        print(database)

     elif cad2.upper() == 'N':
      print('Você entrou como visitante!')

  elif cad.upper() == 'S':

    while True:

      email = input('Digite seu email: ')
      senha = input('Digite sua senha: ')


      for usuario in database:
        if usuario['email'] == email and usuario['senha'] == senha:
          print('Login realizado com sucesso!')
          print('Bem vinda', usuario['nome'])
          return
      print('Email ou senha incorretos! Tente novamente\n')

database = []

def read():
  if not database:
    print('Nenhum usuário cadastrado!')
    return

  print('\nUsuários cadastrados:\n')

  for usuario in database:
    print('ID:', usuario['id'])
    print('Nome:', usuario['nome'])
    print('Email:', usuario['email'])



def update():
  if not database:
    print('Nenhum usuário cadastrado!')
    return
  id_busca = int(input('Digite o ID do usuário que deseja atualizar'))

  for usuario in database:
    if usuario['id'] == id_busca:
      print('Usuário encontrado:', usuario['nome'])

      upd = input('Deseja atualizar alguma informação? S/N')
      if upd.upper() == 'S':
        print('O que deseja atualizar?')
        print('1 - Nome')
        print('2 - Email')
        print('3 - Senha')
        print('4 - Cancelar')
        op = int(input('Escolha uma opção (digite o número correspondente): '))
        if op ==1:
          novo_nome = input('Digite o novo nome: ')
          usuario['nome'] = novo_nome
          print('Nome atualizado com sucesso!')
        if op ==2:
          while True:
            novo_email = input('Digite o novo email')
            if '@' in novo_email and '.' in novo_email:
              break
            else:
              print('Email inválido. Tente novamente.')

          usuario['email'] = novo_email
          print('Email atualizado com sucesso!')

        if op==3:
          nova_senha = input('Digite sua nova senha: ')
          usuario['senha'] = nova_senha
          print('Senha atualizada com sucesso!')
        if op==4:
          print('Operação cancelada.')

        else:
          print('Opção inválida. Tente novamente.')
      return

      print('Usuário não encontrado!')




def delete():
  if not database:
    print('nenhum usuário cadastrado!')
    return

  id_busca = int(input('Digite o ID do usuário que deseja excluir'))
  for usuario in database:
    if usuario['id'] == id_busca:
      print('Usuário encontrado:', usuario['nome'])

      confirm = input('Certeza que deseja excluir esse usuário? S/N')
      if confirm.upper() == 'S':
        database.remove(usuario)
        print('Usuário excluído com sucesso!')
      else:
        print('Operação cancelada.')

  print('Usuário não encontrado.')
  return



print('Boas vindas a LookSocial!\n')

while True:
  print('\nEscolha uma opção:\n')
  print('1 - Cadastrar')
  print('2 - Ver usuários')
  print('3 - Atualizar usuário')
  print('4 - Excluir usuário')
  print('5 - Sair')

  op = input('Escolha uma opção: ')

  if op == '1':
    create()

  elif op =='2':
    read()

  elif op=='3':
    update()

  elif op == '4':
    delete()

  elif op == '5':
    print('Saindo do sistema...')
    break

  else:
    print('Opção inválida.')







```
